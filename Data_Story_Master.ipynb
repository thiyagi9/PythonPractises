```python
#We need to discuss the meaning of each answer. Ex,"Don't know" and "About the same" have same meaning?
happyness_Hness=happyness1['Hness'].replace("Don't know","0")
happyness_Hness=happyness_Hness.replace("About the same","0")
happyness_Hness=happyness_Hness.replace("Somewhat more happy now","1")
happyness_Hness=happyness_Hness.replace("Much more happy now","2")
happyness_Hness=happyness_Hness.replace("Somewhat less happy now","-1")
happyness_Hness=happyness_Hness.replace("Much less happy now","-2")
happyness_Hness.unique()
```

```python
happyness1['Gover'].unique()
```

```python
#We need to discuss the meaning of each answer. Ex,"Don't know" can be neutral point?
happyness_Gover=happyness1['Gover'].replace("Don't know","0")
happyness_Gover=happyness_Gover.replace("Somewhat well","1")
happyness_Gover=happyness_Gover.replace("Very well","2")
happyness_Gover=happyness_Gover.replace("Somewhat badly","-1")
happyness_Gover=happyness_Gover.replace("Very badly","-1")
happyness_Gover.unique()
```

```python
happyness1['Fear'].unique()
```

```python
#We need to discuss the meaning of each answer. Ex,"Don't know" and "Not applicable" have same meaning?
happyness_Fear=happyness1['Fear'].replace("Don't know","0")
happyness_Fear=happyness_Fear.replace("Not applicable - I have already contracted Coronavirus (COVID-19)","0")
happyness_Fear=happyness_Fear.replace("Not applicable - I have already contracted Coronavirus","0")
happyness_Fear=happyness_Fear.replace("I am not very scared that I will contract the Coronavirus (COVID-19)","-1")
happyness_Fear=happyness_Fear.replace("I am not at all scared that I will contract the Coronavirus (COVID-19)","-2")
happyness_Fear=happyness_Fear.replace("I am fairly scared that I will contract the Coronavirus (COVID-19)","1")
happyness_Fear=happyness_Fear.replace("I am very scared that I will contract the Coronavirus (COVID-19)","2")
happyness_Fear.unique()
```
